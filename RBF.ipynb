## the curved RBF decision boundary has the different weight 
## the decision boundary will be closer to the center which has the smaller magnitude that means we need the abs value

```python
import numpy as np
import matplotlib.pyplot as plt

# x range
x = np.linspace(-2, 6, 1000)

# RBF parameters
mu1, mu2 = 0, 4
sigma = 1.2
w1, w2 = 2, -12
b = 0

# Gaussian RBFs
phi1 = np.exp(-((x - mu1)**2) / (2 * sigma**2))
phi2 = np.exp(-((x - mu2)**2) / (2 * sigma**2))

# Output
H = w1 * phi1 + w2 * phi2 + b

# Plot
plt.plot(x, H)
plt.axhline(0, linestyle='--')
plt.scatter([mu1, mu2], [0, 0], marker='x')
plt.xlabel("x")
plt.ylabel("H(x)")
plt.title("RBF Network Output")
plt.show()
```

## Same weight --> the decision boundary is a straight line
## the decision boundary will be closer to the center with smaller weight

```python
# grid
x1 = np.linspace(-2, 6, 400)
x2 = np.linspace(-3, 3, 400)
X1, X2 = np.meshgrid(x1, x2)

# parameters
sigma = 1.0
mu1 = np.array([0, 0])
mu2 = np.array([4, 0])
c = 0.4

# RBFs
phi1 = 5*np.exp(-((X1 - mu1[0])**2 + (X2 - mu1[1])**2) / (2 * sigma**2))
phi2 = 5555*np.exp(-((X1 - mu2[0])**2 + (X2 - mu2[1])**2) / (2 * sigma**2))

# (A)
HA = phi1 - phi2

plt.figure(figsize=(5,4))
plt.contour(X1, X2, HA, levels=[0])   # boundary H_A(x)=0
plt.scatter([0,4], [0,0], marker='x', s=80)
plt.xlim(-2,6)
plt.ylim(-3,3)
plt.title("(A) RBF decision boundary")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(True)
plt.show()
```

## the weight and sigma(width) can control the radius of circle

```python
mu1 = np.array([0, 0])
mu2 = np.array([4, 0])
weight3 = 5
sigma3 = 0.5
phi3 = np.exp(-((X1 - mu2[0])**2 + (X2 - mu2[1])**2) / (2 * sigma3**2))
HB = weight3 *  phi3 - c

plt.figure(figsize=(5,4))
plt.contour(X1, X2, HB, levels=[0])   # boundary H_B(x)=0
plt.scatter([0,4], [0,0], marker='x', s=80)
plt.xlim(-2,6)
plt.ylim(-3,3)
plt.title("(B) RBF decision boundary")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(True)
plt.show()
```
